## Import Package

```python
import pandas as pd
from fastbook import *
from fastai.vision import *
```

## MNIST_SAMPLE Dataset

```python
path = untar_data(URLs.MNIST_SAMPLE)
```

## Cek Path dari Dataset

```python
path.ls()
```

## Memanggil Angka 3 dan 7 pada Path

```python
threes = (path/"train"/'3').ls().sorted()
threes
```

```python
sevens = (path/"train"/'7').ls().sorted()
sevens
```

## Menampilkan Angka 3 dan 7

```python
#visualizing an image of 3
im3_path = threes[1]
im3 = Image.open(im3_path)
im3
```

```python
#visualizing an image of 7
im7_path = sevens[1]
im7 = Image.open(im7_path)
im7
```

## Convert Images to Tensor/Pixel

```python
tensor(im7)[0:28,0:28]
im7_t= tensor(im7)
df = pd.DataFrame(im7_t[0:28,0:28])
df.style.set_properties(**{'font-size': '6pt'}).background_gradient('Greys')
```

```python
tensor(im3)[0:28,0:28]
im3_t= tensor(im3)
df = pd.DataFrame(im3_t[0:28,0:28])
df.style.set_properties(**{'font-size': '6pt'}).background_gradient('Greys')
```
